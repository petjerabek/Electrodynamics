```python
import numpy as np
import matplotlib.pyplot as plt

# Define the scalar field function
def scalar_field(x, y):
    return x**2 + y**2

# Define the vector field function
def vector_field(x, y):
    return 2*x, 2*y

# Create a set of points on the x and y coordinate axes
x_axis = np.linspace(-10, 10, 20)  # 20 points from -10 to 10
y_axis = np.linspace(-10, 10, 20)  # 20 points from -10 to 10

# Generate a 2D grid from the points on the axis
X, Y = np.meshgrid(x_axis, y_axis)

# Compute the scalar field
phi = scalar_field(X, Y)
phi_n = phi / np.max(phi)

# Compute the vector field
Ex, Ey = vector_field(X, Y)

# Plot the "heat map" of the scalar field
cs = plt.contourf(X, Y, phi_n, levels=6)
cbar = plt.colorbar(cs)

# Plot the vector field 
plt.streamplot(X, Y, Ex, Ey, color='tab:blue', linewidth=1.8)


plt.show()

```

```python
def scalar_field(x, y):
    return -x**4 + 4*(x**2-y**2) - 3

def vector_field(x, y):
    return 4*x*(2-x**2), -8*y

x_axis = np.linspace(-2, 2, 100)
y_axis = np.linspace(-2, 2, 100)

X, Y = np.meshgrid(x_axis, y_axis)

phi = scalar_field(X, Y)
phi_n = phi / np.max(phi)

Ex, Ey = vector_field(X, Y)

cs = plt.contourf(X, Y, phi_n, levels=10)
cbar = plt.colorbar(cs)
 
plt.streamplot(X, Y, Ex, Ey, color='tab:blue', linewidth=1.8)

plt.show()
```

# Calculation of the curl

### $V_1 = (2x, 2y)$
* $\frac{\partial E_y}{\partial x} = \frac{\partial (2y)}{\partial x} = 0$
* $\frac{\partial E_x}{\partial y} = \frac{\partial (2x)}{\partial y} = 0$
* **curl:** $0 - 0 = 0$

### $V_2 = (8x - 4x^3, -8y)$
* $\frac{\partial E_y}{\partial x} = \frac{\partial (-8y)}{\partial x} = 0$
* $\frac{\partial E_x}{\partial y} = \frac{\partial (8x - 4x^3)}{\partial y} = 0$
* **curl:** $0 - 0 = 0$

### Explanation
Any vector field field that can be written as a gradient of a scalar potential ($\mathbf{V} = \nabla \phi$) is mathematically guaranteed to have a curl of zero:

$$\nabla \times (\nabla \phi) = 0$$

Physically, this means the fields are conservative and irrotational. The flow lines represent the path of steepest change in the scalar potential and do not form any closed loops or vortices.
