```python
import numpy as np
import matplotlib.pyplot as plt

G = 6.674e-11
M = 1.989e30
m2 = 5.972e24
a = 1.496e11
dt = 86400

t_values = [0]
x_values = [a]
y_values = [0]

v0 = np.sqrt(G * M / a)

Fx = -G * M * m2 * a / (a**3)
Fy = 0

x_one = a + 0.5 * (Fx / m2) * dt**2
y_one = v0 * dt + 0.5 * (Fy / m2) * dt**2

t_values.append(dt)
x_values.append(x_one)
y_values.append(y_one)

while t_values[-1] < 3.15e7:
    r3 = (x_values[-1]**2 + y_values[-1]**2)**1.5
    Fx = -G * M * m2 * x_values[-1] / r3
    Fy = -G * M * m2 * y_values[-1] / r3
    
    x_next = 2 * x_values[-1] - x_values[-2] + (Fx / m2) * dt**2
    y_next = 2 * y_values[-1] - y_values[-2] + (Fy / m2) * dt**2
    
    x_values.append(x_next)
    y_values.append(y_next)
    t_values.append(t_values[-1] + dt)

x_arr = np.array(x_values)
y_arr = np.array(y_values)

plt.plot(x_arr, y_arr)
plt.scatter([0], [0], color='orange')
plt.axis('equal')
plt.title("Planet's circular orbit in 1 year")
plt.xlabel('x position')
plt.ylabel('y position')

plt.show()
```

```python
t_values = [0]
x_values = [a]
y_values = [0]

v0 = 0.6 * np.sqrt(G * M / a)

energy = 0.5 * v0**2 - (G * M / a)
a_new = -G * M / (2 * energy)
period = 2 * np.pi * np.sqrt(a_new**3 / (G * M))

Fx = -G * M * m2 * a / (a**3)
Fy = 0

x_one = a + 0.5 * (Fx / m2) * dt**2
y_one = v0 * dt + 0.5 * (Fy / m2) * dt**2

t_values.append(dt)
x_values.append(x_one)
y_values.append(y_one)

while t_values[-1] < period:
    r3 = (x_values[-1]**2 + y_values[-1]**2)**1.5
    Fx = -G * M * m2 * x_values[-1] / r3
    Fy = -G * M * m2 * y_values[-1] / r3
    
    x_next = 2 * x_values[-1] - x_values[-2] + (Fx / m2) * dt**2
    y_next = 2 * y_values[-1] - y_values[-2] + (Fy / m2) * dt**2
    
    x_values.append(x_next)
    y_values.append(y_next)
    t_values.append(t_values[-1] + dt)

x_arr = np.array(x_values)
y_arr = np.array(y_values)
t_arr = np.array(t_values)

plt.plot(x_arr, y_arr)
plt.scatter([0], [0], color='orange')
plt.axis('equal')
plt.title("Planet's elliptical orbit in 1 year")
plt.xlabel('x position')
plt.ylabel('y position')

plt.show()
```

```python
vx_arr = np.gradient(x_arr, dt)
vy_arr = np.gradient(y_arr, dt)

ke_arr = 0.5 * m2 * (vx_arr**2 + vy_arr**2)
pe_arr = -G * M * m2 / np.sqrt(x_arr**2 + y_arr**2)
e_arr = ke_arr + pe_arr

L_arr = m2 * (x_arr * vy_arr - y_arr * vx_arr)

plt.plot(t_arr, e_arr)
plt.title('Total energy in time')
plt.xlabel('Time')
plt.ylabel('Total energy')
plt.show()

plt.plot(t_arr, L_arr)
plt.title('Angular momentum in time')
plt.xlabel('Time')
plt.ylabel('Angular momentum')
plt.show()
```
